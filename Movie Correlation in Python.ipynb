```python
#import libraries


import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure 
```

```python
#Read the data

df =pd.read_csv('F:\Rahul Sathe\Rahul Sathe Personal\Portfolio Project\Project 4- Python Correlation\movies.csv',encoding='latin-1')
```

```python
#Look at the data 

df.head()
```

```python
#Let's see if there is any missing data 

for col in df.columns:
    pct_missing=np.mean(df[col].isnull())
    print('{}-{}%'.format(col,pct_missing))
```

```python
#Data Type for our columns 

df.dtypes
```

```python
#Change data type of columns

df['budget'] =df['budget'].astype('int64')

df['gross'] =df['gross'].astype('int64')
```

```python
#Create Correct year column 

df['year_correct'] =df['released'].astype('str').str[:4]

df.head()
```

```python
df =df.sort_values(by =['gross'] ,inplace=False ,ascending =False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
# Drop any duplicate 

df['company'].drop_duplicates().sort_values(ascending =False)
```

```python
correlation_matrix =df.corr(method ='pearson')
sns.heatmap(correlation_matrix ,annot =True)
plt.title("Correlation metric for Numeric feature")
plt.xlabel('Movie feature')
plt.ylabel('Movie feature')
plt.show()
```

```python
#Look at Company 

df_numerized =df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype =='object'):
        df_numerized[col_name] =df_numerized[col_name].astype('category')
        df_numerized[col_name] =df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix =df_numerized.corr(method ='pearson')
sns.heatmap(correlation_matrix ,annot =True)
plt.title("Correlation metric for Numeric feature")
plt.xlabel('Movie feature')
plt.ylabel('Movie feature')
plt.show()
```

```python
correleation_mat=df_numerized.corr()

corrleation_pair = correleation_mat.unstack()

corrleation_pair
```

```python
sorted_pairs= corrleation_pair.sort_values()
sorted_pairs
```

```python
High_Corr =sorted_pairs[(sorted_pairs) >.5]
High_Corr
```
